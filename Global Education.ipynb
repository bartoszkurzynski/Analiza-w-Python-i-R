```python
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python

# Download latest version
path = kagglehub.dataset_download("rehan497/global-education")

print("Path to dataset files:", path)
```

```python
df1 = pd.read_csv(os.path.join(path, '1- share-of-the-world-population-with-at-least-basic-education.csv'))
df2 = pd.read_csv(os.path.join(path, '2- learning-adjusted-years-of-school-lays.csv'))
df3 = pd.read_csv(os.path.join(path, '3- number-of-out-of-school-children.csv'))
df4 = pd.read_csv(os.path.join(path, '4- gender-gap-education-levels.csv'))
```

```python
# 1. Merge df1 and df2
merged_df = pd.merge(df1, df2, on=['Entity', 'Code', 'Year'], how='outer')

# 2. Merge the result with df3
merged_df = pd.merge(merged_df, df3, on=['Entity', 'Code', 'Year'], how='outer')

# 3. Merge the result with df4
merged_df = pd.merge(merged_df, df4, on=['Entity', 'Code', 'Year'], how='outer')

# Save the final merged DataFrame
merged_df.to_csv("merged_education_data.csv", index=False)
print(merged_df.head())
```
